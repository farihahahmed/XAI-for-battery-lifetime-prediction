```python
import pandas as pd
import scipy.stats as stats

# Select the three columns from df1
selected_columns = ['LV ActivePower (kW)', 'Wind Speed (m/s)', 'Wind Direction (°)']
df_selected = df1[selected_columns]

# Calculate the statistics using pandas functions
statistics = {
    'Mean': df_selected.mean(),
    'Median': df_selected.median(),
    'Mode': df_selected.mode().iloc[0],
    'Kurtosis': df_selected.kurtosis(),
    'Standard Deviation': df_selected.std(),
    'Maximum': df_selected.max(),
    'Minimum': df_selected.min(),
    'Range': df_selected.max() - df_selected.min()
}

# Create the statistics dataframe
statistics_df = pd.DataFrame(statistics)

# Display the statistics dataframe
print(statistics_df)
```
